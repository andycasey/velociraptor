```python
# What *kinds* of sources did we detect or not detect?
```

```python
consistent_with_single = (unimodal["rv_excess_variance"] == 0)
```

```python
def what_did_we_find(data, consistent_with_single,
                     latex_label_name=None, ax=None, **kwargs):
    if ax is None:
        fig, ax = plt.subplots()
    else:
        fig = ax.figure
    
    kwds = dict(histtype="bar", stacked="True", alpha=0.5)
    kwds.update(kwargs)

    finite = np.isfinite(data)
    ax.hist([
        data[finite * consistent_with_single],
        data[finite * (~consistent_with_single)],
    ], label=("single", "binary"), **kwargs)
    
    legend = plt.legend()
    
    if latex_label_name is not None:
        ax.set_xlabel(latex_label_name)
    
    ax.set_ylabel(r"\textrm{count}")
    
    fig.tight_layout()
    
    return fig
```

```python
fig = what_did_we_find(unimodal["K"], consistent_with_single,
                       latex_label_name=r"$K\,\,/\,\,{\rm km\,s}^{-1}$",
                       bins=30)
```

```python
fig = what_did_we_find(unimodal["P"], consistent_with_single,
                       latex_label_name=r"$P\,\,/\,\,{\rm days}^{-1}$",
                       bins=30)
```

```python
fig = what_did_we_find(unimodal["e"], consistent_with_single,
                       latex_label_name=r"$e$",
                       bins=30)
```

```python
label_names = ("P", "M0", "M1", "K", "v0", "omega", "e")
K = len(label_names)
fig, axes = plt.subplots(1, K, figsize=(4*K, 4))
for k, (ax, label_name) in enumerate(zip(axes, label_names)):
    _ =  what_did_we_find(unimodal[label_name], consistent_with_single,
                         latex_label_name=label_name, bins=30, ax=ax)
    
```
